```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract
import numpy as np
import pandas as pd
import datetime as dt

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 6).all()
print(june)
```

```python
#  3. Convert the June temperatures to a list.
june_temp = list(june)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temp = pd.DataFrame(june_temp, columns=['June temps'])
print(june_temp)
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temp.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.
dec = session.query(Measurement.tobs).filter(extract('month', Measurement.date) == 12).all()
print(dec)
```

```python
# 7. Convert the December temperatures to a list.
dec_temp = list(dec)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 
dec_temp = pd.DataFrame(dec_temp, columns=['December temps'])
print(dec_temp)
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.
dec_temp.describe()
```

```python
# Perform a query to retrieve the data and precipitation scores for June 
june_temp_prcp = session.query(Measurement.tobs, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
june_temp_prcp = list(june_temp_prcp)
june_temp_prcp = pd.DataFrame(june_temp_prcp, columns=['june_temp','precipitation'])
print(june_temp_prcp)
```

```python
june_temp_prcp.describe()
```

```python
june_temps=june_temp_prcp['june_temp']
june_prcps=june_temp_prcp['precipitation']
```

```python
plt.subplots(figsize=(18, 6))
plt.scatter(june_temps,
            june_prcps, s=300, c="skyblue",
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.5, label="June")
plt.title("June Temperature and Precipitation")
plt.ylabel("Precipitation")
plt.xlabel("Temperature")
plt.grid(True)
plt.legend()
```

```python
# Perform a query to retrieve the data and precipitation scores for December
dec_temp_prcp = session.query(Measurement.tobs, Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()
dec_temp_prcp = list(dec_temp_prcp)
dec_temp_prcp = pd.DataFrame(dec_temp_prcp, columns=['dec_temp','precipitation'])
```

```python
print(dec_temp_prcp)
```

```python
dec_temp_prcp.describe()
```

```python
dec_temp=dec_temp_prcp['dec_temp']
dec_prcp=dec_temp_prcp['precipitation']
```

```python
plt.subplots(figsize=(18, 6))
plt.scatter(dec_temp,
            dec_prcp, s=300, c="coral",
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.5, label="December")
plt.title("December Temperature and Precipitation")
plt.ylabel("Precipitation")
plt.xlabel("Temperature")
plt.grid(True)
plt.legend()
```

```python
# List the stations and the counts in descending order for June
session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).filter(extract('month', Measurement.date) == 6).order_by(func.count(Measurement.station).desc()).all()
```

```python
# List the stations and the counts in descending order for December
session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).filter(extract('month', Measurement.date) == 12).order_by(func.count(Measurement.station).desc()).all()
```

```python
# Using the station id, calculate the lowest, highest, and average temperature most active station for June
session.query(Measurement.station,func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).filter(extract('month', Measurement.date) == 6).filter(Measurement.station == 'USC00519281').all()
```
